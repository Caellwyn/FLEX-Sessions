```python
#__SOLUTION__
def check_data(features, target, categorical=False):
    """
    This function takes a dataframe of features and a series or array of targets and 
    prints the shape, datatypes, number of missing values, and class values of each
    
    """
    
    print('First 5 Rows of Features')
    display(features.head(5))
    print('\n')
    print('Shape of Features: ', features.shape)
    print('\n')
    print('Features and Data Types')
    print(features.info())
    print('\n')
    print('Missing Data in Features')
    print(features.isna().sum())
    print('\n')
    print('First 5 Rows of the Target')
    display(target.head())
    print('\n')
    print('Target Data Type: ', target.dtype)
    print('\n')
    print('Missing Labels in Target', target.isna().sum())
    print('\n')
    if categorical:
        print('Target Class Balance')
        print(target.value_counts())
    else:
        print('Target Minimum Value: ', np.min(target))
        print('\n')
        print('Target Maximum Value: ', np.max(target))
```

```python
check_data(df.drop('target', axis=1), df.target)
```
